# Python et Java

## Jython

https://www.jython.org

Une des premières solutions pour marier Java et Python. 
Implémentation de Python sur la JVM. 

Python 2.7 seulement. Dernière version : 2017. 

Le développement de Jython 3.5 est +/- au point mort par manque de sponsors. 

## JPype

http://www.jpype.org/

Pont entre CPython et la machine virtuelle Java.

Donne accès aux bibliothèques Java depuis Python. 

## Py4J

https://www.py4j.org/

Même principe que JPype. 

> Py4J is more focused than JPype. For example, it does not attempt to link Python threads to Java threads. JPype also uses JNI to communicate with the JVM while Py4J uses plain old sockets, which are more portable in practice.

Pas la solution la plus performante :

> In terms of performance, Py4J has a bigger overhead than both of the previous solutions because it relies on sockets, but if performance is critical to your application, accessing Java objects from Python programs might not be the best idea :-) 

Dernière version : octobre 2018. 

## Pyjnius

https://pyjnius.readthedocs.io/

Développé dans le cadre du projet python-for-android, par le projet Kivy. 

> We found others libraries such as JPype or Py4j, but we weren’t happy with the design and the usability. 

> We have a mapping between Java/Python type, native arrays, support for methods with multiple signatures, and a lot more. We are using Cython + JNI internally. The performance cost is minimal.

https://blog.kivy.org/2012/08/pyjnius-accessing-java-classes-from-python/

Dernière version : décembre 2019. 

## Jep

https://pypi.org/project/jep/

https://github.com/ninia/jep

Python 2.7 et 3.3 ou plus. 

Approche différente : permet d’embarquer l’interpréteur CPython dans la JVM (via JNI).

Donne accès aux bibliothèques Python depuis Java. Les bibliothèques Python pures sont toutes supportées, mais pas forcément celles qui ont des extensions C. 

Dernière version : août 2019. 

## JCC

Développé dans le cadre du projet PyLucene. 

http://lucene.apache.org/pylucene/jcc/index.html

Permet de générer une extension CPython en code C++ qui emballer des classes Java et les accéder via JNI. Nécessite un compilateur C++. 

Dernière version : septembre 2019.

## GraalVM

https://www.graalvm.org

Il existe des implémentations de Python et de Java sur GraalVM, et des mécanismes sont fournis pour faire communiquer les deux : https://github.com/graalvm/graalpython/blob/master/doc/POLYGLOT.md

Attention, encore très expérimental !
